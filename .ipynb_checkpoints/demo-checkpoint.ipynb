```python
import pandas as pd
df = pd.read_csv("cardio_train.csv",sep=";")
df
```

```python
df2 = df.copy()
```

```python
df2["age_years"] = (df["age"] / 365).round(1)
df2["age_years"]
```

```python
df2 = df2[(df2["height"]>=120) & (df2["height"]<=220)]
df2 = df2[(df2["weight"]>=35) & (df2["weight"]<=180)]
```

```python
df2["bmi"] = df2["weight"] / ((df2["height"]/100)**2)
```

```python
df2 = df2[(df2["ap_hi"]>=50) & (df2["ap_hi"]<=250) &
            (df2["ap_lo"]>=30) & (df2["ap_lo"]<=200) &
            (df2["ap_hi"]>df2["ap_lo"])]
```

```python
df2.drop('age',axis=1)
```

```python
df2.to_csv('cleaned_data.csv',index=False)
```

```python
data = pd.read_csv('cleaned_data.csv')
```

```python
data.head()
```

```python
import sklearn
sklearn.__version__
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['id','age','cardio'],axis=1)
y = data['cardio']

X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=40)
```

```python
y.value_counts()
data.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=40
)
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

preds = model.predict(X_test)

print("Accuracy:", accuracy_score(Y_test, preds))
print(classification_report(Y_test, preds))
print(confusion_matrix(Y_test, preds))
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [8, 12, None]
}

grid = GridSearchCV(RandomForestClassifier(), params, cv=3, n_jobs=-1)
grid.fit(X_train, Y_train)

print(grid.best_params_)
```
